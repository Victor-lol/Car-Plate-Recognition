```python
import cv2
import matplotlib.pyplot as plt
import imutils
import numpy as np
from skimage import measure
path = "plate.png"
img = cv2.imread(path)
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
s = hsv[:, :, 2]
plt.imshow(img)
```

```python
plt.imshow(hsv)
```

```python
plt.imshow(s)
```

```python
ret, thresh = cv2.threshold(s, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU) #threshold
plt.imshow(thresh)
```

```python
cnts = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)#contors find
cnts = imutils.grab_contours(cnts)
c = max(cnts, key=cv2.contourArea)
x, y, w, h = cv2.boundingRect(c)
out = img[y:y+h, x:x+w, :].copy()
plt.imshow(out)
```

```python
gray = cv2.cvtColor(out,cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
```

```python
plt.imshow(thresh)
```

```python
labels = measure.label(thresh, connectivity=2, background=0)
charCandidates = np.zeros(thresh.shape, dtype="uint8")
```

```python
plt.imshow(labels)
```

```python
plt.imsave('labeled.png', labels)
```

```python
plt.imsave('cropped.png', out)
```

```python
labelMask = np.zeros(thresh.shape, dtype="uint8")
for y in range(len(labels)):
    for x in range(len(labels[y])):
        if labels[y][x]> 2:
            labelMask[y][x] = 255
```

```python
plt.imshow(labelMask)
```

```python
plt.imsave('labelMask.png', labelMask)
```

```python
labelMask2 = np.transpose(labelMask)
plt.imshow(labelMask2)
```

```python
flagList = []
flag = False
for r in labelMask2:
    if np.count_nonzero(r)!=0:
        flag = True
    else:
        flag = False
    flagList.append(flag)  
```

```python
imList = []
sigle_ch = []
for i in range(len(flagList)):
    if flagList[i] == True:
        sigle_ch.append(labelMask2[i])
    else:
        if sigle_ch:
            imList.append(sigle_ch)
            sigle_ch = []
```

```python
for i in range(len(imList)):
    imList[i] = np.transpose(imList[i])
```

```python
for i in imList:
    plt.imshow(i)
```

```python
final_image_list = []
for i in imList:
    flagList = []
    flage = False
    for j in i:
        if np.count_nonzero(j)!=0:
            flag = True
        else:
            flag = False
        flagList.append(flag)
    sigle_ch = []
    for k in range(len(flagList)):
        if flagList[k] == True:
            sigle_ch.append(i[k])
        else:
            if sigle_ch:
                final_image_list.append(sigle_ch)
                sigle_ch = []
```

```python
plt.imshow(final_image_list[0])
```

```python
final_image_list = np.array(final_image_list)
```

```python
for i in range(len(final_image_list)):
    final_image_list[i] = np.array(final_image_list[i])
    for j in range(len(final_image_list[i])):
        final_image_list[i][j] = np.array(final_image_list[i][j])
```

```python
test_set = []
```

```python
for i in range(len(final_image_list)):
    final_image_list[i] = cv2.resize(final_image_list[i], dsize=(28, 28), interpolation=cv2.INTER_CUBIC)
plt.imshow(final_image_list[0])
```

```python
test = []
for i in range(len(final_image_list)):
    final_image_list[i]=final_image_list[i].flatten()
    test.append(final_image_list[i])
```
