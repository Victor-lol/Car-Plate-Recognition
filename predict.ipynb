```python
# k fold cross validation for model selection
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

kf = KFold(n_splits=9,shuffle=True,random_state=4)
# each [a,b] represnets a components and b neighbours
prams = [[200,1],[200,3],[200,5],[100,1],[100,3],[100,5],[400,1],[400,3],[400,5]]
i = 0

acc = []
for train_index, test_index in kf.split(X):
  # get each fold in cross validation
  X_train, X_test = X[train_index], X[test_index]
  y_train, y_test = y[train_index], y[test_index]


  # apply PCA for each fold
  pca = PCA(n_components=prams[i][0])
  X_train = pca.fit_transform(X_train)
  X_test = pca.transform(X_test)

  # fit knn for each fold
  knn = KNeighborsClassifier(n_neighbors=prams[i][1])
  knn.fit(X_train, np.ravel(y_train))
  y_pred = knn.predict(X_test)

  print("Components: ",prams[i][0]," Neighbors: ",prams[i][1]," Accuracy:",metrics.accuracy_score(y_test, y_pred))
  acc.append(metrics.accuracy_score(y_test, y_pred))
  i = i+1
```

```python
# get the best model with the whole dataset
index = acc.index(max(acc))
pca = PCA(n_components=prams[index][0])
X = pca.fit_transform(X)  # transform the data into new space

knn = KNeighborsClassifier(n_neighbors=prams[index][1])
knn.fit(X,np.ravel(y))

# to do: take the subimages from part 2 and apply it here
# y_pred = knn.predict(X_new) # X_new is the input matrix with each row as an observation
```
